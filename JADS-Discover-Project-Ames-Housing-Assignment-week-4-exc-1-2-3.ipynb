### Defining a function to explore a dataframe and return categorized lists of featuresÂ¶

```python
def expl_df(df_input, name):
    
    """ Prints analysis of dataframe 'df_input' with name 'name' and returns lists of numeric, categorical, complete,
        incomplete features, features with many zeros and features with high skew > 0.75. """

    print('\n\033[1mDataframe analysis for dataframe:  ',name,'\033[0m\n')

    # Print dimensions of dataframe
    num_rows = df_input.shape[0]
    num_cols = df_input.shape[1]
    print('Dataframe',name,'has', num_rows, 'rows and', num_cols,'columns\n')

    # Check which features are numeric and which features are categorical 
    num_features = list(df_input.select_dtypes(include=['int64','float64']).columns)
    cat_features = list(df_input.select_dtypes(include=['object', 'category']).columns)
    print('Dataframe',name,'has', len(num_features) , 'numeric features and', len(cat_features),'categorical features\n')

    # Check which features have missing values and might need attention and print number of features
    count_values_feature=df_input.count()

    complete_features = list(count_values_feature[count_values_feature == num_rows].index)
    incomplete_features = list(count_values_feature[count_values_feature < num_rows].index)
    print('Dataframe',name,'has',len(complete_features), 'complete features and', len(incomplete_features),'incomplete features\n')
    
    # Check which features have more than 50% zero values and print number of features
    zeroes_feature = (df_input[num_features] == 0).sum() / num_rows
    high_zero_features = list(zeroes_feature[zeroes_feature > 0.5].index)
    print('Dataframe',name,'has',len(high_zero_features), 'features with more than 50% zeroes\n')

    # Calculate skew for each numeric feature and select features with skew>0.75 
    calc_skew_feature = df_input[num_features].skew()
    skew_features = list(calc_skew_feature[calc_skew_feature > 0.75].index)
    print('Dataframe',name,'has',len(skew_features), 'numeric features with a skew > 0.75\n')
    
    return num_features, cat_features, incomplete_features, complete_features, high_zero_features, skew_features
```

### Define a function to create space-efficient way to plot 1D histograms for all remaining numeric variables

```python
#
# 1D histogram plots for all numeric features in dataframe without features in list
# 'list_of_features_to_be_dropped' sorted by feature name 
#

def plt_1D_histograms_for_dataframe_numeric_features(dataframe, list_of_features_to_be_dropped):

    """ Plot 1D histogram plots for all numerical features in pandas dataframe sorted by feature name """

    # Create list with Numerical Features of dataframe
    numeric_features = dataframe.select_dtypes(include=['int64','float64']).columns

    # Drop features as per list_of_features_to_be_dropped and convert to list
    if len(list_of_features_to_be_dropped) > 0:
        numeric_features = list(numeric_features.drop(list_of_features_to_be_dropped))
    
    # Calculate number of plots per column
    num_plt_cols = len(numeric_features)
    first_num_plt_cols = num_plt_cols - int(num_plt_cols/2)

    # Create subplots in 2 columns
    fig, axes = plt.subplots(first_num_plt_cols, 2, figsize=(16,first_num_plt_cols))

    # Create subplots in first column
    i=0
    for column in numeric_features[0:first_num_plt_cols]:
        dataframe[column].dropna(inplace=True)
        bins = np.histogram(dataframe[column], bins=100)[0]
        bins = np.expand_dims(bins, axis=0)
        axes[i,0].imshow(bins, cmap='YlOrRd', aspect='auto')
        axes[i,0].get_xaxis().set_visible(False)
        axes[i,0].get_yaxis().set_visible(False)
        axes[i,0].set_title(str(column), loc='left')
        i = i+1

    # Create subplots in second column
    i=0
    for column in numeric_features[first_num_plt_cols:num_plt_cols]:
        dataframe[column].dropna(inplace=True)
        bins = np.histogram(dataframe[column], bins=100)[0]
        bins = np.expand_dims(bins, axis=0)
        axes[i,1].imshow(bins, cmap='YlOrRd', aspect='auto')
        axes[i,1].get_xaxis().set_visible(False)
        axes[i,1].get_yaxis().set_visible(False)
        axes[i,1].set_title(str(column), loc='left')
        i = i+1

    plt.tight_layout()
    plt.show()
    return
```

```python
#
# Plot 1D histograms for all numeric features with defined function 
#
plt_1D_histograms_for_dataframe_numeric_features(df,[])
```

### Replace Missing Values 

```python
#
# Replace missing values with "missing" in the categorical features and with median in numeric features
#
for column in incomp_feat:
    if (str(df[column].dtype) == "category") | (str(df[column].dtype) == "object"):
        print('Fixing categorical feature:\t', column, '\tby replacing missing values with label "missing"')

        try:
            df[column] = df[column].cat.add_categories('missing') # add "missing" to categories before replacing 
        except:
            continue # "missing" was already a category for categorical features

        df[column].fillna('missing', inplace = True) 
    elif (str(df[column].dtype) == "int64") | (str(df[column].dtype) == "float64"):
        print('Fixing numeric feature:', column,'\tby replacing missing values with median:\t', df[column].median())
        df[column].fillna(df[column].mean(), inplace=True)
    
    else:
        print('\nUnexpected datatype for feature\n', column)
```

### Investigating Multicollinearity between numeric features

See https://www.analyticsvidhya.com/blog/2020/03/what-is-multicollinearity/ for example

```python
#
# Import libraries for VIF, Mapping and define a function to calculate the variance_inflation_factor to 
# determine multicollinearity
#
# VIF starts at 1 and has no upper limit
# VIF = 1, no correlation between the independent feature and other features
# VIF exceeding 5 or 10 indicates high multicollinearity between a feature and other features
#
from statsmodels.stats.outliers_influence import variance_inflation_factor
from collections import Mapping

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["Feature"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
#
# Prepare dataframe with all numeric features without SalePrice and calcuate VIZ matrix
#

# Switch off warnings to avoid annoying warnings
warnings.filterwarnings('ignore')

# Create new feature list 'viz_feat' for VIZ calculations
viz_feat = num_feat.copy()

# Remove SalePrice from viz_feat list if it is present
if 'SalePrice' in viz_feat:
    viz_feat.remove('SalePrice')
    
# Create new dataframe X 
X = df[viz_feat]

# Calculate and sort variance inflation factor for all selected features
vif_matrix = calc_vif(X).sort_values(by=['VIF'], ascending=False)

# Force 'unscientific' formatting of VIF 
pd.options.display.float_format = '{:.2f}'.format

# Print dataframe without distracting index. N.B. VIF > 5 indicates high multi-collinearity
print(vif_matrix.to_string(index=False))

# Store all feaure names with VIF > 5.0 in list 'viz_feat'
vif_feat = vif_matrix[vif_matrix['VIF'] > 5.0].iloc[:,0]

# Switch on warnings again
warnings.filterwarnings('default')
```
